```python
df["reservation_status_date"]=pd.to_datetime(df["reservation_status_date"])
```

```python
df.drop(["agent","company"],axis=1,inplace=True)
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
df['adr'].plot(kind='box')
```

```python
df=df[df['adr']<5000]
```

```python
df.describe()
```

```python
is_canceld=df["is_canceled"].value_counts(normalize=True)
is_canceld
```

```python
df['is_canceled']=df['is_canceled'].astype(str)
```

```python
plt.figure(figsize=(5, 6))
sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Pastel1')
plt.title('Canceld rates in City hotel and Resort hotel', fontweight='bold', size=20)
plt.show()
```

its very important note

```python
city_hotel=df[df['hotel']=='city Hotel'] 
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
print(city_hotel['is_canceled'].value_counts(normalize=True))
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
print(resort_hotel['is_canceled'].value_counts(normalize=True))
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
resort_hotel
```

```python
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel
```

```python
plt.figure(figsize=(20, 8), facecolor='#FFFFFF') 
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=30)


plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel', color='green')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel', color='blue')

plt.legend(fontsize=20)
plt.show()
```

```python

df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")


sns.countplot(x='month', hue='is_canceled', data=df, palette='Set2')

plt.title('Reservation Status by Month', fontsize=18, fontweight='bold')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Number of Reservations', fontsize=14)


plt.legend(title='Reservation Status', labels=['Not Canceled', 'Canceled'])

plt.tight_layout()
plt.show()
```

```python
cancelled_data= df[df['is_canceled']=='1']
top_10_country = cancelled_data['country'].value_counts().head(10)
plt.figure(figsize=(8, 8))
plt.title('Top 10 countries with reservation canceled', color="black")
plt.pie(top_10_country, autopct='%.2f', labels=top_10_country.index)
plt.show()
```
